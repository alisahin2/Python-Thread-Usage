```python
import threading
```

```python
print("Example-1")
```

```python
def loop1():
    for i in range(1,20000):
        print("1", end="")
        
def loop2():
    for i in range(1,5000):
        print("2", end="")
```

```python
loop1()
loop2()
```

```python
thread1 = threading.Thread(target = loop1)
thread2 = threading.Thread(target = loop2)

thread1.start()
thread2.start()
```
